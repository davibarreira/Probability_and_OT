```julia
Random.seed!(1)
N = []
p = plot()
n = 5
 = 3*randn(n)
data = []
for i=1:n
    v = rand(Normal([i],1),1000)
    if i == 1
        data = v
    else
        data = vcat(data,v)
    end
    
    push!(N,Normal([i],1))
    p = plot!(N[i],linewidth = 2,legend = false)
end
p2 = density(data,label="Euclidean Average",grid=true)
p2 = plot!(Normal(mean(),1),legend = true, label="Barycenter")
s = plot(p,p2,layout=2,size=(800,300))
savefig(s, "./images/Barycenter.png")
s
```
